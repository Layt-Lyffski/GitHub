```python
#filters //"~" negate the condition
fav_countries = ['Germany', 'Poland', 'United Kingdom', 'Ireland', 'Austria']
filt = df["Country"].isin(fav_countries)
```

```python
df['Hobbyist'] = df['Hobbyist'].replace({'Yes' : True, 'No' : False})
```

## Grouping and Aggregating
Aggregation compile plural data into singel Result (mean, median)
### Grouping
    • Split
    • Apply Funcitn
    • Combine results

```python
country_grp = df.groupby(['Country']) #groupby([]) allow to groupby a col
country_grp.get_group('United States') #get_group() get the values that can be stored under col 'Country'
#display all grouped all rows that fit to the grouping condition
```

```python
filt = df['Country'] == 'United States' #create filter
df.loc[filt]['SocialMedia'].value_counts() #apply filt on SocialMedia and count all result with .value_counts()
#display the aggregated results
```

```python
#if we combine the two above command into one: //normalize give you percentage result
country_grp['SocialMedia'].value_counts(normalize=True).loc[['Poland', 'Germany']] #worth it use it 
```

```python
country_grp['ConvertedComp'].agg(['median', 'mean']).loc[fav_countries]
```

```python
#str can be used only on series of data mean not on df and after apply(func e.g. lambda x:) with func you can set the protocol how should work the code
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).loc[fav_countries]
```

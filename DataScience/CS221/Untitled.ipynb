## Content-based Recommender Systems

##### Notation
$n_u$ = no. users<br>
$n_m$ = no moves<br>
$r(i,j)$ = 1 if user $j$ has rated movie $i$<br>
$y^{(i,j)}$ = rating given by user $j$ to move $i$ (defined only if $r(i,j)$ = 1)<br>
$\theta^{(j)}$ = parameter vector for user $j$<br>
$x^{(i)}$ = feature vector for movie $i$<br>
###### Problem formulation
For user $j$, movie $i$, predicted rating: $(\theta^{(j)})^T(x^{(i)})$<br>
m^{(j)} = no. of movies rated by user $j$ <br>
<br>
#### Optimization Objective:
To learn $\theta^{(j)}$ (parameter for user $j$):
$$\binom{min}{\theta^{(j)}} \frac{1}{2}\sum_\limits{i:r(i,j)=1}((\theta^{(j)})^T x^{(i)}-y^{(i,j)})^2 + \frac{\lambda}{2}\sum_\limits{k=1}^n (\theta_k^{(j)})^2$$

To learn $\theta^{1},\theta{(2)},...,\theta^{n_u}$:
$$\binom{min}{\theta^{(1)},...,\theta^{n_u}} \frac{1}{2}\sum_\limits{j=1}^{n_u}\sum_\limits{i:r(i,j)=1}((\theta^{(j)})^T x^{(i)}-y^{(i,j)})^2 + \frac{\lambda}{2}\sum_\limits{j=1}^{n_u}\sum_\limits{k=1}^n (\theta_k^{(j)})^2$$<br>

### Algorithm

<img src="pic\algo.png"><br><br>
1. Initialize $x^{(1)},...,x^{(n_m)}, \theta^{(1)},...,\theta^{(n_u)}$ to small random values.<br>
2. Minimizue $J(x^{(1)},...,x^{(n_m)},\theta^{(1)},...,\theta^{n_u})$ using gradient descent (or an advanced opt. algorithm). E.g. for every $j = 1,...,n_u, i = 1,...,n_m:$<br>

$$x_k^{(i)}:= x_k^{(i)}-\alpha \sum_\limits{Rj:r(i,j)=1}((\theta^{(j)}^T x^{(i)}-y^{(i,j)})\theta_k^{(j)}+\lambda x_k^{(i)}$$

$$\theta_k^{(i)}:= \theta_k^{(i)}-\alpha \sum_\limits{j:r(i,j)=1}((\theta^{(j)}^T x^{(i)}-y^{(i,j)})\theta_k^{(j)}+\lambda \theta_k^{(i)}$$ 

where the in () is the $\frac{\delta}{\delta x^{(i)}}J(...) and x\in\mathbb{R}^n; \theta\in\mathbb{R}^n$<br>
